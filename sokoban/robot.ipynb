```python
import pygame
import sys
from SokobanPuzzle import SokobanPuzzle
from Node import Node

game_map1 = [
    ['O', 'O', 'O', 'O','O','O','O'],
    ['O', 'O', ' ', ' ','O','O','O'],
    ['O', 'O', ' ', ' ','O','O','O'],
    ['O', 'O', ' ', '*',' ',' ','O'],
    ['O', 'O', 'B', 'O','B',' ','O'],
    ['O', ' ', 'S', 'R','S',' ','O'],
    ['O', ' ', ' ', ' ',' ','O','O'],
    ['O', 'O', 'O', ' ',' ','O','O'],
    ['O', 'O', 'O', 'O','O','O','O']
]


pygame.init()

width, height = len(game_map1[0]) * 100, len(game_map1) * 100
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Game Example")


player_img = pygame.image.load('assets/player.png')
player_img = pygame.transform.scale(player_img, (100, 100))

blue_square_image = pygame.image.load('assets/obstacle.png')
blue_square_image = pygame.transform.scale(blue_square_image, (100, 100))


full_storage_image = pygame.image.load('assets/full_storage.png')
full_storage_image = pygame.transform.scale(full_storage_image, (100, 100))

box_image = pygame.image.load('assets/box.png')
box_image = pygame.transform.scale(box_image, (100, 100))

storage_image = pygame.image.load('assets/storage.png')
storage_image = pygame.transform.scale(storage_image, (100, 100))


sokoban_puzzle = SokobanPuzzle(grid=game_map1)
init_node = Node(sokoban_puzzle)
states = init_node.getStates()
states.insert(0, game_map1)


running = True
current_state_index = 0
state_display_delay = 1000  
last_state_change_time = pygame.time.get_ticks()
font = pygame.font.Font(None, 74)  



while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    
    current_time = pygame.time.get_ticks()
    if current_time - last_state_change_time > state_display_delay:
        
        if current_state_index < len(states) - 1:
            current_state_index += 1
            game_map1 = states[current_state_index]
            last_state_change_time = current_time
        else:
                screen.fill((0,0,0))  
                text = font.render("Game Over", True, (255, 0 ,0))  
                text_rect = text.get_rect(center=(width // 2, height // 2))
                screen.blit(text, text_rect)  

                
                pygame.display.flip()

                
                waiting = True
                while waiting:
                    for event in pygame.event.get():
                        if event.type == pygame.QUIT:  
                            pygame.quit()
                            sys.exit()
                        elif event.type == pygame.KEYDOWN:  
                            if event.key == pygame.K_r:  
                                waiting = False

    
    screen.fill((0, 0, 0))

    
    for j in range(len(game_map1)):
        for i in range(len(game_map1[j])):
            if game_map1[j][i] == 'O':
                screen.blit(blue_square_image, (100 * i, 100 * j))
            elif game_map1[j][i] == 'S':
                screen.blit(storage_image, (100 * i, 100 * j))
            elif game_map1[j][i] == 'B':
                screen.blit(box_image, (100 * i, 100 * j))
            elif game_map1[j][i] == 'R':
                screen.blit(player_img, (100 * i, 100 * j))
            elif game_map1[j][i] == '.':
                screen.blit(storage_image, (100 * i, 100 * j))
                screen.blit(player_img, (100 * i, 100 * j))
            elif game_map1[j][i] == '*' : 
                screen.blit(full_storage_image, (100 * i, 100 * j))


            
    
    
    pygame.display.flip()

pygame.quit()
sys.exit()
```

```python
import pygame
import sys
from interface import Interface

# Initialize Pygame
pygame.init()

# Screen settings
screen_width, screen_height = 800, 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Pathfinding Game")

# Font settings
font = pygame.font.Font(None, 50)
small_font = pygame.font.Font(None, 36)

# Define game states
STATE_WELCOME = "welcome"
STATE_MAP_SELECTION = "map_selection"
STATE_ALGORITHM_SELECTION = "algorithm_selection"
STATE_SOLVING = "solving"

# Initial state
current_state = STATE_WELCOME
selected_map = None
selected_algorithm = None

# Map options and algorithm choices
maps = ["Map 1", "Map 2", "Map 3"]
algorithms = ["A*", "BFS"]

# Instantiate the interface and pass necessary parameters
interface = Interface(screen, font, small_font, maps, algorithms)

# Clock for limiting the frame rate
clock = pygame.time.Clock()

# Main loop
running = True
while running:
    # Fill the screen with black before each frame
    screen.fill((0, 0, 0))

    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN and event.button == 1:  # Left click
            mouse_pos = pygame.mouse.get_pos()

            # Check the current state and transition accordingly
            if current_state == STATE_WELCOME:
                play_button_rect = interface.show_welcome_screen()
                if play_button_rect.collidepoint(mouse_pos):
                    current_state = STATE_MAP_SELECTION  # Move to map selection
            elif current_state == STATE_MAP_SELECTION:
                map_buttons = interface.show_map_selection_screen()
                for button_rect, map_name in map_buttons:
                    if button_rect.collidepoint(mouse_pos):
                        selected_map = map_name
                        current_state = STATE_ALGORITHM_SELECTION  # Move to algorithm selection
            elif current_state == STATE_ALGORITHM_SELECTION:
                algorithm_buttons = interface.show_algorithm_selection_screen()
                for button_rect, algo in algorithm_buttons:
                    if button_rect.collidepoint(mouse_pos):
                        selected_algorithm = algo
                        current_state = STATE_SOLVING  # Start solving

    # Render based on the current state
    if current_state == STATE_WELCOME:
        interface.show_welcome_screen()
    elif current_state == STATE_MAP_SELECTION:
        interface.show_map_selection_screen()
    elif current_state == STATE_ALGORITHM_SELECTION:
        interface.show_algorithm_selection_screen()
    elif current_state == STATE_SOLVING:
        interface.start_solving(selected_algorithm)

    # Update the display
    pygame.display.flip()

    # Limit the frame rate
    clock.tick(30)  # 30 FPS

pygame.quit()
sys.exit()
```
